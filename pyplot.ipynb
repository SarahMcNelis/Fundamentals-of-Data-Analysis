## Matplotlib

[Matplotlib](https://en.wikipedia.org/wiki/Matplotlib) is a comprehensive plotting library for the Python programming language which is used for data visualisation. It was origionally written by [John D. Hunter](https://en.wikipedia.org/wiki/John_D._Hunter) who was an American neurobiologist (1968–2012). The idea of Matplotlib came about during his reseach into neurobiology. He needed a resource to visualise the electrocorticography (ECoG) data of epilepsy patients. Since then, Matplotlib has an [active development community](https://github.com/matplotlib) lead by [Michael Droettboom](https://github.com/mdboom) and [Thomas Caswell](https://github.com/tacaswell). The official matplotlib website also has a [discourse page](https://discourse.matplotlib.org/) which allows the community and devolopers to interact and discuss problems and solutions. 

***

## Pyplot

[Pyplot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.html) is a module in Matplotlib. Pyplot has 137 functions and commands that make matplotlib behave like [MATLAB](https://www.mathworks.com/products/matlab.html). Matlab is another programming and numeric computing platform used by millions of engineers and scientists worldwide. Similar to matlab, pyplot is a state-based interface. This is different to an object orientated approach. Instead of assigning a piece of code to an object, pyplot defines and stores the current state using pyplot commands and functions. This allows pyplot to keep track of the desired state and automatically updates it once a command is called. Therefore, it is fair to say that pyplot is a convenient way for users to generate high quality plots with minimal effort.  

***

## NumPy
 
[Numpy](https://numpy.org/doc/stable/user/whatisnumpy.html) is an open source library in python. It is used when working with arrays and can perform a wide variety of mathematical operations. It is more efficent and more compact then using python lists. 
[The numpy random package](https://numpy.org/doc/stable/reference/random/index.html) is a module in numpy that is used to generate random numbers. These are pseudo-random numbers which means they are not truly random because every random sequence is actually determinded by an inital value. By [default](https://www.w3schools.com/python/ref_random_seed.asp) the bit generator is none and therefore uses the current system time as the initial value. This is why the return value will differ each time the code is run.  
[Seed](https://www.geeksforgeeks.org/random-seed-in-python/) is a funciton within the numpy random module that is used to set the initial value when generating random numbers. Once a seed value is set then each time that piece of code is executed it generates the same random numbers. The advantage of this is that when running the code it doesn't matter what the current time is or which machine you are using because the ramdom data will remain unchanged.

# Horizontal Bar Plot


[Offical documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.barh.html)


https://matplotlib.org/stable/gallery/lines_bars_and_markers/barh.html#sphx-glr-gallery-lines-bars-and-markers-barh-py 

***

## Parameters:
***

- ***y***: determines where the bars are positioned. In the below code cell the [numpy arange function](https://numpy.org/doc/stable/reference/generated/numpy.arange.html) is used to return an evenly spaced values within a defined interval. The default start value for np.arange is 0 and the stop value is set at the lenght of places which is 5. The plot below displays evenly spaced cities on the y-axis. 


- ***width***: sets the value of the bars. In this example, the width is the variable 'popularity'. This variable uses [numpy random rand function](https://www.geeksforgeeks.org/numpy-random-rand-python/) to create an array with five values and multiply those values by three plus five thousand. 


- ***height***: this parameter sets the height of each bar. If not selected the default value for height is 0.8. 


- ***left***: determines the left point on the x-axis. By default this is 0. 


- ***align***: refers to the alignment of the bars on the y-axis. This example sets align to 'center' which centers the bars on the y positions. The other option here is 'edge' which aligns the bottom edges of the bars with the y positions. This is set to center by default. 

<br>

## Pyplot Functions used:
***

### yticks
This function is used to set ticks on the y-axis. This plot sets the [yticks](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.yticks.html) to the length of places which is five. Then labels are set for these ticks to the actual place names. This allows for a clear understanding of what each bar represents. 


### gca
The [gca](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.gca.html) function is part of the pyplot package and it allows users to get the current axes on a figure. In this example, the gca function is used to invert the y-axis so that it appears in the order of the list on the plot from top to bottom. 


### xlabel
[This function](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xlabel.html) sets a text label on the x-axis. Users can set the location of this label on the figure by passing in the loc parameter. The default location for xlabels is center as can be seen in the plot below. 


### title
[plt.title](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.title.html) adds a title to the figure. The default location is center, however, another can be specified by using the loc parameter for this as well. 


### show
[Pyplot.show](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.show.html) is what calls and displays a plot. However, this notebook has already run the magic command `%matplotlib inline` which automatically displays the plot at the end of every code cell by default. Nonetheless, the show function presents the plot in a more clean cut way. 


### savefig
This [function](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.savefig.html) is used to save a figure. This is done to access the figure outside of the notebook.

<br>

```python
# Changing figure style. 
plt.style.use('bmh')

# Fixing random state for reproducibility.
np.random.seed(5000)

# Example data.
places = ('London', 'New York', 'Sydney', 'Tokyo', 'Paris')
# Array of 5 values.
y = np.arange(len(places)) 
# Generating ramdon data for popularity.
popularity =  3 + 5000 *np.random.rand(len(places)) 

# Creating the horizontal barplot.
plt.barh(y, popularity)
# Set ticks and labels.
plt.yticks(ticks=y, labels=places)
# Invert the labels.
plt.gca().invert_yaxis()
plt.xlabel('Popularity')
plt.title('What is the most popular city?')
plt.show()
#plt.savefig('img/horizontal_barplot.png')
```

# Scatter Plot

[Offical documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html)


https://matplotlib.org/stable/gallery/shapes_and_collections/scatter.html#sphx-glr-gallery-shapes-and-collections-scatter-py

***

## Parameters of a Scatter Plot
***

- ***x,y***: the first step to creating a scatter plot is setting the data positions. There must be x data and y data. In the below code cell, numpy random rand is used to generate an array for x and y containing random values. These two arrays are then passed into the scatter plot in order to create 80 marker points on the plot. 


- ***s***: this is an optional parameter that is used to change the size of the marker points. In other words, if a list or array of size is passed into the argument, then it returns a scatter plot with different marker sizes. In the plot below, s is set the the variable area. This variable contaions a list of 80 random float values which determine the size/area of each marker point in this scatter plot. If not selected the default marker size is rcParams['lines.markersize'] ** 2. 


- ***c***: is another optional parameter which passes in a color or list of colors. The variable colors was set to generate 80 random colors to use in the plot. 


- ***marker***: can be set by the instance of the class or using the word for that marker. A list of possible markers can be found [here](https://matplotlib.org/stable/api/_as_gen/matplotlib.markers.MarkerStyle.html#matplotlib.markers.MarkerStyle). The default marker is 'o' which can be seen in the plot below.


- ***cmap***: is used if the c option is an array of floats. The [colormap](https://matplotlib.org/stable/api/_as_gen/matplotlib.colors.Colormap.html#matplotlib.colors.Colormap) converts an array of floats to the RGBA color that it represents. 


- ***norm***: is used if c is an array of floats. It scales 'c' in the range 0-1 and then maps it into the colormap. The default is none which [normalises](https://matplotlib.org/stable/api/_as_gen/matplotlib.colors.Normalize.html#matplotlib.colors.Normalize) the data. 


- ***vmin, vmax***: are used alongside the default for norm. In other words, they use the minimun and maximum values to normalise the 'c' values in order to create the colormap. 


- ***alpha***: is the blending value of the markers. This value is a float between 0 (transparent) and 1 (opaque). This is set to 0.5 in the below code cell which allows for slight transparenty in the plot. 


- ***linewidths***: determines the lonewidth or the marker edges and is represented as a float. The default linewidth is 1.5. 


- ***edgecolors***: will default at face value. In other words the edge color will be the same as the color selected in 'c'. If none is selected the edgecolor will have no patch boundary. If a color is selected it will change the color of the edge of every marker point. 

 
- ***plotnonfinite***: is a boolean option which defaults to false. If True, the points are displayed with the bad colormap color. 

<br>

```python
# Changing figure style. 
plt.style.use('fivethirtyeight')

# Fixing random state.
np.random.seed(19680801)

# create random floats for x and y.
N = 80
x = np.random.rand(N) 
y = np.random.rand(N) 
# setting 80 random colors.
colors = np.random.rand(N) 
# creating 80 random sizes/area of each marker.
area = (30 * np.random.rand(N))**2  

# Call and display scatterplot. 
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
#plt.savefig('img/scatterplot.png')
```

# Histograms

[Offical documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html)


https://matplotlib.org/stable/gallery/pyplots/pyplot_text.html#sphx-glr-gallery-pyplots-pyplot-text-py

***

## Parameters of a Histogram
***

- ***x***: is the input value which can be an array or sequence of arrays. In the code cell below, the variable which returns 10000 random float values multiplied by the sum of mean and sigma. 


- ***bins***: can be an integer or sequence of integers. It sets the number of equal bin widths in the range. A bin sequence defines the bin edges. This example sets the bin width to 50.


- ***range***: if this parameter is set it refers to the lower and upper range of the bins. In this plot there is no range as the defaul is none. 


- ***density***: is a boolean option which if true it returns a probability density. This is done by dividing the bin's raw count by the sum of counts multiplied by the bin width. The formula for this is ``(density = counts / (sum(counts) * np.diff(bins)))``. The reason is to ensure that the area under the histogram integrates to 1. 


- ***weights***: can be an array the same shape as 'x' where only each value in 'x' contributes its weight to the bin count. However, if density is true, the weights are normalised. 


- ***cumulative***: this is a boolean option or can also be -1. The default for this is false. If true the cumlative value of the histogram gives the counts for each bin on the y-axis plus all bins for smaller values of the response variable. The last bin gives the total number of datapoints. If density is also set to true then the plot is normalized  so that the last bin equals 1. If cumulative is an integer less than 0 then the direction of accumulation is reversed. However, if the density is true with a negative cumulative value then the plot is normalised so that the first bin equals 1. 


- ***bottom***: refers to the bottom of each bin. In other words, the bottom starts on the left edge of the bin and ends on the right edge. Therefore, this parameter can move the bottom of the bins or each bin independently. If not selected it defaults to 0. 


- ***histtype***: defaults to bar histogram. However, other types can be chosen by selecting this parameter. For example, barstacked; where bars are stacked on top of each other, step; which draws a lineplot which is unfilled, stepfilled; which draws a filled lineplot. This plot uses the default option of a bar type histogram. 


- ***align***: determines where the bars are alligned on the histogram. It can be either left, mid or right of the bin edges. The default option is mid. 


- ***orientation***: can be either vertical (default) or horizontal. If horizontal a barh will be used instead. 


- ***rwidth***: the relative width is automatically computed by default. However, if selected is gives the fraction of the bin width. 


- ***log***: is another boolean option which defaults to false. If true then a log scale will be selected for the axis. 


- ***color***: if unselected it uses a standard line color for the plot. 


- ***label***: will add a string to match multiple datasets on the plot. 


- ***stacked***: is a boolean option which default to false where data is arranged side by side as opposed to true where the data is stacked on top of each other. 

<br>

## Pyplot Functions used:
***

### xlabel
This [function](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xlabel.html) sets the label for the x-axis to 'Smarts'. The location of this label can be left, center, right. If not selected it defaults at center. 


### ylabel
[ylabel](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.ylabel.html) displays the label 'Probability' on the y-axis. The label location can be selected as bottom, center, top. Center is the default option. 


### title
This pyplot function sets a [title](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.title.html) to the plot which by default displays in the center. However, the loc parameter can be used to display either left or right. 


### text
[Pyplot text](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.text.html) adds text to an x, y location. This example sets text to display the mu and sigma on the x-axis at 60 and the y-axis at 0.25. 


### xlim
This [function](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.text.html) sets the x-axis limit to 40 on the left and 160 on the right. 


### ylim
Similar to xlim, [ylim](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.ylim.html) sets the limits on the y-axis. In the example below the bottom limit is set to 0 and the top to 0.03. 


### grid
[Pyplot grid](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.grid.html) is a boolean option which if True, it displays the plot with grid lines. 

<br>

```python
# Changing figure style. 
plt.style.use('ggplot')

# Fixing random state.
np.random.seed(19680801)

# Set mean and standard deviation.
mu, sigma = 100, 15
# Create an array of 10000 random values for mu and sigma. 
x = mu + sigma * np.random.randn(10000)

# The histogram of the data.
plt.hist(x, 50, density=True, alpha=0.8)
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.xlim(40, 160)
plt.ylim(0, 0.03)
plt.grid(True)
plt.show()
#plt.savefig('img/histogram.png')
```

# Subplots 

[Offical documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html)


https://matplotlib.org/stable/gallery/lines_bars_and_markers/categorical_variables.html#sphx-glr-gallery-lines-bars-and-markers-categorical-variables-py

***
 
 [Subplots](https://matplotlib.org/stable/gallery/subplots_axes_and_figures/subplots_demo.html) is a function in the pyplot package which creates a figure and a set of subplots. In other words, it is a simple way of creating one figure and axes with multiple subplots with just one call.

## Parameters of a Subplot
***

- ***nrows, ncols***: is a integer which refers to the rows or columns of the subplot. The default value for this is 1. This example sets 1 row with 3 columns. 


- ***sharex, sharey***: is a boolean option which defaults to false. If true, all the x-axis or y-axis data will be shared among all subplots. If false then none will be independent of each other. The other options within this parameter are row; which means only each subplot row will share an x or y-axis, and col; which will only share each column. In the example below True has been selected which means all of the x and y-axis will be shared. 


- ***squeeze***: is also a boolean parameter. The default for this is True which means that extra dimensions are extracted from the returned array of axes. 


- ***subplot_kw***: is a dict which adds an axes to a figure in order to create each subplot.


- ***gridspec_kw***: is also a dict which is used to create the grid the subplots are put on. 


- *****fig_kw***: is used to pass any additional arguments to the pyplot figure call.  

<br>

## Pyplot Functions used:
***

First the subplot function is used to set the [figure](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figure.html?highlight=pyplot%20figure#matplotlib.pyplot.figure) and [axs](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.axes.html).  


### axs.bar
Creates a [bar plot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html) at value 0 on the row which is the first column. 

### axs.scatter
Uses the [pyplot scatter function](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html) to create the second subplot at value 1 on the row which is the second column. 

### axs.plot
The [pyplot plot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html) function is used which plots a y verus x line. This is set at value 2 on the row which is the third column. 

### fig.subtitle
This uses the [pyplot subtitle](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.suptitle.html) function to give the figure a title. 

<br>

```python
# Changing figure style. 
plt.style.use('classic')

# Give fruit random values. 
data = {'apple': 10, 'orange': 15, 'lemon': 5, 'lime': 20}
names = list(data.keys())
values = list(data.values())

# Create subplot with 1 row and 3 columns which shares data.
fig, axs = plt.subplots(1, 3, figsize=(14, 6), sharey=True)
# Set first ax to a bar plot.
axs[0].bar(names, values)
# Set second ax to a scatter plot.
axs[1].scatter(names, values)
# Set third ax to a lineplot
axs[2].plot(names, values)
fig.suptitle('Categorical Plotting')
plt.show()
#plt.savefig('img/subplot.png')
```
